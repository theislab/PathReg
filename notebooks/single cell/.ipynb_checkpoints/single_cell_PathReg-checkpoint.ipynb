```python
class initial_position(nn.Module):
    
    def __init__(self, dim, nhidden):
        super(initial_position, self).__init__()
        
    def forward(self, x): 
        
        x0 = torch.mean(x,axis=0)
        for g in range (x.shape[2]):
            zscore = (x[...,g] - x[...,g].mean()) / torch.sqrt(x[...,g].var())
            zscore = torch.where(torch.isnan(zscore), torch.zeros_like(zscore), zscore)
            x0[:,g] = torch.mean(x[...,g][zscore<3])
        return x0  

class ODEBlock(nn.Module):

    def __init__(self, odefunc, dim):
        super(ODEBlock, self).__init__()
        self.odefunc = odefunc
        self.dim = dim
        
    def set_times(self,times):
        self.integration_times = times
        
    def forward(self, x):
        integrated = odeint(self.odefunc, x, self.integration_times, rtol=args['tol'], atol=args['tol'], method = args['method'])
        out = torch.empty(len(self.integration_times), 1, self.dim)
        for i in range(len(self.integration_times)):
            out[i] = integrated[i][0,:self.dim]
        return out

    @property
    def nfe(self):
        return self.odefunc.nfe

    @nfe.setter
    def nfe(self, value):
        self.odefunc.nfe = value


def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)


class L0_MLP(nn.Module):
    def __init__(self, input_dim, layer_dims=(100, 100), N=50000, beta_ema=0.999,
                 weight_decay=0., lambas=(1., 1., 1.), local_rep=False, temperature=2./3.):
        super(L0_MLP, self).__init__()
        self.layer_dims = layer_dims
        self.input_dim = input_dim
        self.N = N
        self.beta_ema = beta_ema
        self.weight_decay = self.N * weight_decay
        self.lambas = lambas

        layers = []
        for i, dimh in enumerate(self.layer_dims):
            inp_dim = self.input_dim if i == 0 else self.layer_dims[i - 1]
            droprate_init, lamba = 0.2 if i == 0 else 0.5, lambas[i] if len(lambas) > 1 else lambas[0]
            if i<len(self.layer_dims)-2:
                layers += [L0Dense(inp_dim, dimh, droprate_init=droprate_init, weight_decay=self.weight_decay,
                               lamba=lamba, local_rep=local_rep, temperature=temperature)]
                layers += [nn.ELU(alpha=1, inplace=False)]
            else:
                layers += [L0Dense(inp_dim, dimh, droprate_init=droprate_init, weight_decay=self.weight_decay,
                               lamba=lamba, local_rep=local_rep, temperature=temperature)]
        self.output = nn.Sequential(*layers)

        self.layers = []
        for m in self.modules():
            if isinstance(m, L0Dense):
                self.layers.append(m)
        
        self.nfe = 0
            
        if beta_ema > 0.:
            print('Using temporal averaging with beta: {}'.format(beta_ema))
            self.avg_param = deepcopy(list(p.data for p in self.parameters()))
            if torch.cuda.is_available():
                self.avg_param = [a.cuda() for a in self.avg_param]
            self.steps_ema = 0.
        self.nfe = 0

       
    def forward(self, t, x):  
        self.nfe += 1
        return self.output(x)
    
    def regularization(self):
        regularization = 0.
        for layer in self.l0_layers:
            regularization += - (1. / self.N) * layer.regularization()
        if torch.cuda.is_available():
            regularization = regularization.cuda()
        return regularization

    def get_exp_flops_l0(self):
        expected_flops, expected_l0 = 0., 0.
        for layer in self.l0_layers:
            e_fl, e_l0 = layer.count_expected_flops_and_l0()
            expected_flops += e_fl
            expected_l0 += e_l0
        return expected_flops, expected_l0

    def update_ema(self):
        self.steps_ema += 1
        for p, avg_p in zip(self.parameters(), self.avg_param):
            avg_p.mul_(self.beta_ema).add_((1 - self.beta_ema) * p.data)

    def load_ema_params(self):
        for p, avg_p in zip(self.parameters(), self.avg_param):
            p.data.copy_(avg_p / (1 - self.beta_ema**self.steps_ema))

    def load_params(self, params):
        for p, avg_p in zip(self.parameters(), params):
            p.data.copy_(avg_p)

    def get_params(self):
        params = deepcopy(list(p.data for p in self.parameters()))
        return params
```

```python
def array_tensor(X,T,X0):
    tX = []
    tT = []
    tX0 = []
    for i in range(len(X)):
        data = X[i]
        t = T[i]
        x0 = X0[i]
        data_dim = data.shape[1]
        samp_ts = torch.empty(data.shape[0]).float()   
        for i in range (data.shape[0]):
            samp_ts[i]= torch.tensor(float(t[i]))
        z0= torch.empty((x0.shape[0],1,data_dim)).float().to(device)
        z = torch.empty((data.shape[0], 1, data_dim)).float().to(device)
        for j in range(data_dim):
            for i in range(x0.shape[0]):    
                z0[i,0,j] = torch.tensor(float(x0[i,j]))
            for i in range(data.shape[0]):
                z[i,0,j] = torch.tensor(float(data[i, j]))
        tX.append(z)
        tT.append(samp_ts)
        tX0.append(z0)
    return tX,tT,tX0
```

```python
def _remove_cells(data,time,idx1,idx2):
    for i in range(len(data)):
        data[i] = data[i][idx1:idx2]
        time[i] = time[i][idx1:idx2]
        assert(len(data[i])==len(time[i]))
    return data,time

def _nan2zero(x):
    return np.where(np.isnan(x), np.zeros_like(x), x)

def _mean(data):
    for i in range(len(data)):
        data[i] = (data[i] - np.min(data[i],axis=0))/ (np.max(data[i],axis=0) - np.min(data[i],axis=0))
        data[i] =_nan2zero(data[i])
    return data
```

```python
def _initial_cond(data,n):
    init = []
    for i in range(len(data)):
        init.append(data[i][:n,:])
    return init
```

# Model

```python
## model
nhidden = 100
data_dim = data_a_12.shape[1]
feature_layers = [initial_position(data_dim, nhidden), 
                  ODEBlock(
                      L0_MLP(input_dim=data_dim, layer_dims=(nhidden,nhidden,data_dim), N=150),
                      data_dim
                  )]
model = nn.Sequential(*feature_layers).to(device)
loss_func = nn.MSELoss()
print(model)
```

```python
def PathReg(model):
    for i, layer in enumerate(model[1].odefunc.layers):
        if i ==0:
            WM = torch.abs(layer.sample_weights_ones())
        else:
            WM = torch.matmul(WM,torch.abs(layer.sample_weights_ones()))
    return torch.mean(torch.abs(WM))
```

```python
def L1(model):
    for i, layer in enumerate(model[1].odefunc.layers):
        if i ==0:
            WM = torch.abs(layer.weights)
        else:
            WM = torch.matmul(WM,torch.abs(layer.weights))
    return torch.mean(torch.abs(WM))
```

```python
def train(lam):
    filename = './models/PathReg/'
    feature_layers = [initial_position(data_dim, nhidden), 
                  ODEBlock(
                      L0_MLP(input_dim=data_dim, layer_dims=(nhidden,nhidden,data_dim), N=150),
                      data_dim
                  )]
    model = nn.Sequential(*feature_layers).to(device)
    optimizer = optim.Adam(model.parameters(), lr=args['lr'])
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', 0.5, 20,1e-6,'rel',min_lr=1e-16,verbose=True)
    loss_func = nn.MSELoss()
    
    if os.path.isfile(filename+'single_cell_'+str(lam)+'.pth'):  # LOAD CHECKPOINT
        checkpoint = torch.load(filename+'single_cell_'+str(lam)+'.pth')
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        epoch = checkpoint['epoch']
        loss = checkpoint['loss']
        losses = checkpoint['losses']

    else:
        epoch = 1
        losses = []

    

    itr_arr = np.empty(args['niters'])
    loss_arr = np.empty(args['niters'])
    nfe_arr = np.empty(args['niters'])
    time_arr = np.empty(args['niters'])

    #training
    for itr in range(epoch, args['niters'] + 1):
        
        lambda_l1 = lam
        lambda_path = 0.1*lambda_l1
        
        feature_layers[1].nfe = 0
        optimizer.zero_grad()

        # forward in time and solve ode
        for i in range(args['n_samples']):
            loss = 0
            model[1].set_times(tT[i*3])
            pred_z = model(tX0[i*3].to(device))
            loss += loss_func(pred_z.to(device), tX[i*3].to(device))

            model[1].set_times(tT[i*3+1])
            pred_z = model(tX0[i*3+1].to(device))
            loss += loss_func(pred_z.to(device), tX[i*3+1].to(device))

            model[1].set_times(tT[i*3+2])
            pred_z = model(tX0[i*3+2].to(device))
            loss += loss_func(pred_z.to(device), tX[i*3+2].to(device))

            # compute regularizers
            path_reg = PathReg(model)
            l1_reg = L1(model)
            
            loss += (lambda_l1 * l1_reg + lambda_path * path_reg)

            loss.backward()
            optimizer.step()
            scheduler.step(loss)

        #make arrays
        itr_arr[itr-1] = itr
        loss_arr[itr-1] = loss
        nfe_arr[itr-1] = feature_layers[1].nfe

        print('Iter: {}, running MSE: {:.9f}'.format(itr, loss), end="\r")
        torch.save({
            'epoch': itr,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': loss,
            'losses': losses
            }, filename+'single_cell_'+str(lam)+'.pth')
    return model
              
lambdas = [0.01]
for lam in lambdas:
    print('TRAIN WITH LAMBDA', lam)
    model = train(lam)
```

```python
def get_loss(model,test):
    preds = []
    loss_func = nn.MSELoss()
    losses = []
    if test:
        r0 = 1
        r1 = 5
    else:
        r0 = 0
        r1 = 1
    for i in range(r0,r1): 
        loss = 0 
        model[1].set_times(tT[i*3])
        pred_z = model(tX0[i*3])
        loss += loss_func(pred_z.to(device), tX[i*3].to(device))
        preds = preds + [pred_z]

        model[1].set_times(tT[i*3+1])
        pred_z = model(tX0[i*3+1])
        loss += loss_func(pred_z.to(device), tX[i*3+1].to(device))
        preds = preds + [pred_z]

        model[1].set_times(tT[i*3+2])
        pred_z = model(tX0[i*3+2])
        loss += loss_func(pred_z.to(device), tX[i*3+2].to(device))
        preds = preds + [pred_z]
        losses.append(loss)
    return sum(losses) / len(losses)
```

```python
def get_sparsity(model):
    
    for i, layer in enumerate(model[1].odefunc.layers):
        if i ==0:
            WM = torch.abs(layer.sample_weights_ones())
        else:
            WM = torch.matmul(WM,torch.abs(layer.sample_weights_ones()))
    WM = WM.T.cpu().detach().numpy()
    
    # FEATURE SPARSITY
    WM[np.abs(WM)<1e-5] = 0.
    feat_spar = (WM == 0).sum() / (WM.shape[0] * WM.shape[1])
    
    # MODEL SPARSITY
    for i, layer in enumerate(model[1].odefunc.layers):
        if i ==0:
            all_weights = torch.abs(layer.sample_weights()).flatten()
        else:
            all_weights = torch.cat((all_weights,torch.abs(layer.sample_weights()).flatten()))
    model_spar = (torch.abs(all_weights)<1e-5).sum() / all_weights.shape[0]
    
    # SPURIOUS FEATURE SPARSITY
    df_WM = pd.DataFrame(WM,index=genes_12,columns = list(genes_12))
    TFs = pd.read_csv('../../data/data/TFs_Targets/tf_targets_baseGRN_human_matrix.csv.gz',sep=',',index_col=0,compression='gzip')
    TFs_shared = np.intersect1d(genes_21,TFs.index)
    genes_shared = np.intersect1d(genes_21,TFs.columns)
    temp = np.array(df_WM.loc[:,~df_WM.columns.isin(TFs_shared)])
    spur_feat_spar = (temp<1e-5).sum() / (temp.shape[0]*temp.shape[1])

    return feat_spar, model_spar, spur_feat_spar
```

```python
filename = './models/PathReg/'
lambdas = [1e-2]
all_mse = []
all_mse_test = []
all_feat_spar = []
all_model_spar = []
all_spur_feat_spar = []
for lam in lambdas:
    nhidden = 100
    data_dim = data_a_12.shape[1]
    feature_layers = [initial_position(data_dim, nhidden), 
                  ODEBlock(
                      L0_MLP(input_dim=data_dim, layer_dims=(nhidden,nhidden,data_dim), N=150),
                      data_dim
                  )]
    model = nn.Sequential(*feature_layers).to(device)
    
    loss_func = nn.MSELoss()
    print('looking for model ', filename+'single_cell_'+str(lam)+'.pth')
    if os.path.isfile(filename+'single_cell_'+str(lam)+'.pth'):  # LOAD CHECKPOINT
        checkpoint = torch.load(filename+'single_cell_'+str(lam)+'.pth', map_location=torch.device(device))
        model.load_state_dict(checkpoint['model_state_dict'])
    else:
        print('couldnt find the model')
    mse = get_loss(model,test=False)
    all_mse.append(mse)
    
    mse_test = get_loss(model,test=True)
    all_mse_test.append(mse_test)
    
    feat_spar, model_spar, spur_feat_spar = get_sparsity(model)
    all_feat_spar.append(feat_spar)
    all_model_spar.append(model_spar)
    all_spur_feat_spar.append(spur_feat_spar)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = lambdas
y1 = torch.tensor(all_mse).cpu().detach().numpy().tolist()
y1a = torch.tensor(all_mse_test).cpu().detach().numpy().tolist()
y2a = torch.tensor(all_feat_spar).cpu().detach().numpy().tolist()
y2b = torch.tensor(all_model_spar).cpu().detach().numpy().tolist()
y2c = torch.tensor(all_spur_feat_spar).cpu().detach().numpy().tolist()

fig, ax1 = plt.subplots()

ax1.set_xlabel('Lambda')
ax1.set_ylabel('MSE Loss', color='gray')
ax1.plot(x, y1, color='gray', label='Train Loss')
ax1.plot(x, y1a, color='gray', linestyle='dotted', label='Test Loss')
ax1.tick_params(axis='y', labelcolor='gray')
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('Sparsity', color='olive')  # we already handled the x-label with ax1
ax2.plot(x, y2b, color='olive', linestyle='solid', label='Model Sparsity')
ax2.plot(x, y2a, color='olive', linestyle='dotted', label='Feature Sparsity',marker='o')
ax2.plot(x, y2c, color='olive', linestyle='dashed', label='Spurious Feature Sparsity')
ax2.tick_params(axis='y', labelcolor='olive')
fig.legend(bbox_to_anchor=(0.56, 0.58))
plt.xscale("log")
plt.title("PathReg")
plt.savefig('lambdas_sc_pathreg.pdf')
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
